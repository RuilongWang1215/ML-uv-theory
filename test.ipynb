```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
argon = pd.read_csv('data/real_ufrac_argon.csv')
methane = pd.read_csv('data/real_ufrac_methane.csv')
water = pd.read_csv('data/real_ufrac_water.csv')
```

```python
methane['density'] = methane['density']/1000
argon['density'] = argon['density']/1000
water['density'] = water['density']/1000
```

```python
fig = plt.figure(figsize=(18, 4))
plt.subplot(1,3,1)
#x is density, y is temperature, the color of the point is delta_phi, and add a color bar
scatter = plt.scatter(methane['density'], methane['delta_phi'], c=methane['temperature'], cmap='viridis')
color_bar = plt.colorbar(scatter)
color_bar.set_label('Temperature')
plt.xlabel('Density')
plt.ylabel('$\\Delta \\phi$')
plt.title('Methane')

plt.subplot(1,3,2)
scatter = plt.scatter(water['density'], water['delta_phi'], c=water['temperature'], cmap='viridis')
color_bar = plt.colorbar(scatter)
color_bar.set_label('Temperature')
plt.xlabel('Density')
plt.ylabel('$\\Delta \\phi$')
plt.ylim(-2,2)
plt.title('Water')

plt.subplot(1,3,3)
scatter = plt.scatter(argon['density'], argon['delta_phi'], c=argon['temperature'], cmap='viridis')
color_bar = plt.colorbar(scatter)
color_bar.set_label('Temperature')
plt.xlabel('Density')
plt.ylabel('$\\Delta \\phi$')
plt.title('Argon')

plt.show()
```

```python
print('name| water | methane | argon\n----|-------|---------|------')
print('density range| {:.2f}-{:.2f} | {:.2f}-{:.2f} | {:.2f}-{:.2f}'.format(water['density'].min(), water['density'].max(), methane['density'].min(), methane['density'].max(), argon['density'].min(), argon['density'].max()))
print('temperature range| {:.2f}-{:.2f} | {:.2f}-{:.2f} | {:.2f}-{:.2f}'.format(water['temperature'].min(), water['temperature'].max(), methane['temperature'].min(), methane['temperature'].max(), argon['temperature'].min(), argon['temperature'].max()))
print('$\\Delta \\phi$ range| {:.2f}-{:.2f} | {:.2f}-{:.2f} | {:.2f}-{:.2f}'.format(water['delta_phi'].min(), water['delta_phi'].max(), methane['delta_phi'].min(), methane['delta_phi'].max(), argon['delta_phi'].min(), argon['delta_phi'].max()))
print('number of data points| {} | {} | {}'.format(len(water), len(methane), len(argon)))
```
